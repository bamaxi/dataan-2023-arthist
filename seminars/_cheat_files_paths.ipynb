# Как найти, где моя тетрадка, и куда класть файлы, с которыми работаем?

**путь** — местоположение файла или папки на компьютере  
* **абсолютный путь** — полный путь, доступен из любого другого местоположения на компьютере.  
      На Windows вида `D:\Documents\Teach\hse-dataan-2023-arthist\repo` (в начале буква диска)
* **относительный путь** — местоположение относительно текущей папки.  
      Везде может писаться как `/seminars/data/`, например.  
      Может включать в записи сокращения:  
          `.` — текущая папка, `/seminars/data/` = `./seminars/data/`,  
          `..` — родительская папка, `../misc/` папка `misc`, находящаяся на одном уровне с моей текущей папкой (дочь родителя=сестра)

### 2. Куда новые тетрадки и файлы?

2а. **В найденную папку** и нужно помещать новые тетрадки и файлы, с которыми работаем!

При структуре как на картинке, когда тетрадка, из которой хочу открыть файлы (`09-files.ipynb`) лежит в той же папке, что и файлы (`first_text.txt` и `gismeteo.csv` например),
    очень легко их открыть...
* через `with open("ЗДЕСЬ_ИМЯ_ФАЙЛА")`
* или `pd.read_csv("ЗДЕСЬ_ИМЯ_ФАЙЛА")` (`pd` это из `import pandas as pd`)

Например, `pd.read_csv("gismeteo.csv")`

С другой стороны, файлы можно положить в подпапку (здесь `data`) относительно той, где ваша тетрадка (`09-py-files.ipynb`)

Открыть файл `BikeData.csv` можно как `pd.read_csv("data/BikeData.csv")`

И когда мы открывали как `pd.read_csv("data/BikeData.csv")`, и когда как `pd.read_csv("gismeteo.csv")` мы использовали **относительные пути**.
